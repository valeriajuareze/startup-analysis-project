```python
import pandas as pd
```

```python
startup_df = pd.read_csv("Resources/startup_data.csv")
```

```python
startup_df.state_code.value_counts()
```

```python
startup_df.drop(columns=["closed_at", "Unnamed: 6", "status", "id", "object_id", "name", "state_code.1","city", "state_code", "category_code", "Unnamed: 0", "latitude", "longitude", "zip_code"], axis=1, inplace=True)
```

```python
startup_df.age_first_milestone_year.value_counts()
```

```python
startup_df[startup_df.age_first_milestone_year.isnull()].labels.value_counts()
```

```python
startup_df["reached_milestone"] = startup_df.age_first_milestone_year.map(lambda x: 0 if pd.isnull(x) else 1)
```

```python
startup_df.head()
```

```python
# startup_df.dropna(inplace=True)
```

```python
startup_df.columns
```

```python
startup_df.founded_at = pd.to_datetime(startup_df.founded_at)
startup_df.first_funding_at = pd.to_datetime(startup_df.first_funding_at)
startup_df.last_funding_at = pd.to_datetime(startup_df.last_funding_at)
```

```python
startup_df.founded_at.map(lambda x: x.year)
```

```python
startup_df["founded_first_funding__days_difference"] = startup_df.first_funding_at - startup_df.founded_at
startup_df["first_last_funding_days_difference"] = startup_df.last_funding_at - startup_df.first_funding_at
```

```python
startup_df.founded_first_funding__days_difference = startup_df.founded_first_funding__days_difference.astype('timedelta64[D]').astype(int)
startup_df.first_last_funding_days_difference = startup_df.first_last_funding_days_difference.astype('timedelta64[D]').astype(int)
```

```python
startup_df.drop(columns=["founded_at", "first_funding_at", "last_funding_at"], axis=1, inplace=True)
```

```python
corr_matrix=startup_df.corr()
corr_matrix["labels"].sort_values(ascending=False).plot(kind="bar")
```

```python
startup_df.to_csv("Resources/startup_data_cleaned.csv")
```
